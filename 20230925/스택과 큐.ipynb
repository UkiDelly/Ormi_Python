# Stack과 Queue


```python
# 다음중 Stack이 아닌것은?
# 이런 문제는 정보처리기사에만 나옵니다.
# 코테랑은 관련 없고, 개념을 제대로 이해하고 있는지 보는 것입니다.
# 다음중 스택이 아닌것은?
# 1, 2, 3, 4, 5 순서로 데이터가 들어갑니다.
# 1. [5, 4, 3, 2, 1](스택 O)
# 2. [3, 2, 1, 5, 4](스택 O)
# 3. [1, 2, 3, 4, 5](스택 O)
# 4. [2, 1, 4, 3, 5](스택 O)

# 1. 입구가 [0]에 있기 때문에 데이터를 넣으면 [1] -> [2,1] ... -> [5,4,3,2,1]
# 2. 입구가 [0], [3]에 있고 1~3은 [0]입구에, 4,5는 [3]입구에 넣어서 [1,_,_,4,_] -> [2,1,_,5,4] -> [3,2,1,5,4]
# 3. 입구가 마지막 인덱스에 있어서 [1,2,3,4,5]
# 4. 입구가 [0], [2], [4]에 있고 1,2은 [0]입구에, 3,4는 [2]입구에, 5는 [4]입구에 넣어서 -> [ 1,_,3,_,5 ] -> [2,1,4,3,5]


# 또는
# 1. 데이터:[1,2,3,4,5], stack=[] -> 데이터:[1,2,3,4], stack=[5] ... -> 데이터:[] , stack =[5,4,3,2,1]
# 2. 데이터:[1,2,3] & [4,5], stack=[] -> 데이터:[1,2] & [4,5], stack=[3] -> 데이터:[1] & [4,5], stack=[3,2] ... -> 데이터:[] & [], stack=[3,2,1,5,4]
# 3. 데이터:[5,4,3,2,1] ,stack=[] -> 데이터:[5,4,3,2] ,stack=[1] .... -> 데이터:[] ,stack=[1,2,3,4,5]
# 4. 데이터: [1,2] & [3,4] & [5], stack=[] -> 데이터: [1] & [3,4] & [5], stack=[2] -> 데이터: [] & [3,4] & [5], stack=[2,1] -> 데이터: [] & [3] & [5], stack=[2,1,4] ... -> 데이터: []&[]&[],stack=[2,1,4,3,5]
```
